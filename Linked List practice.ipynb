```python
# insert in ll recursively:
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
def print_ll(ll):
    while ll is not None:
        print(str(ll.data)+"->",end='')
        ll=ll.next
    print('None')
    return 

def length(ll):
    count=0
    while ll is not None:
        count+=1
        ll=ll.next
    return count
    
def insert_rec(head,i,data):
    if i <0 or i >length(head):
        return head
    if i ==0:  # base case is this
        new_node=Node(data)
        new_node.next=head
        return new_node
    
    
    small_head=insert_rec(head.next,i-1,data)
    head.next=small_head
    return head

def traverse_input():
    input_list=[int(ele) for ele in input().split()]
    head=None
    tail=None
    for i in input_list:
        if i==-1:
            break
        new_node=Node(i)
        if head is None:
            head=new_node
            tail=new_node
        else:
            tail.next=new_node
            tail=new_node
    return head

head=traverse_input()
print_ll(head)
new_ll=insert_rec(head,2,6) # somewher mid
print_ll(new_ll)
new_ll_ll=insert_rec(new_ll,0,8)  # in starting
print_ll(new_ll_ll)
new_ll_ll_ll=insert_rec(new_ll_ll,length(new_ll_ll),99)# in end
print_ll(new_ll_ll_ll)
```

```python
# reversing linked list  using recursion:  O (n**2)
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

def print_ll(ll):
    while ll is not None:
        print(str(ll.data)+"->",end='')
        ll=ll.next
        
    print("None")
    
    return

def length(ll):
    count=0
    while ll is not None:
        count+=1
        ll=ll.next
    return count

def reverse(ll):
    
    if ll is None:
        return None
    
    if ll.next is None: # agr ek hei ele hai linked list mei
        return ll
    
    small_head=reverse(ll.next)
    curr=small_head
    while curr.next is not None:
        curr=curr.next
    curr.next=ll # yeh smallhead wali ll ke sath humne input wali ll ka head join kar diya 
    ll.next=None  # aur finally ek None
    
    return small_head # return smallhead na ki ll woh toh origional wali hai na

def traverse_input():
    input_list=[int(ele) for ele in input().split()]
    head=None
    tail=None
    for i in input_list:
        if i==-1:
            break
        new_node=Node(i)
        if head is None:
            head=new_node
            tail=new_node
        else:
            tail.next=new_node
            tail=new_node
    return head

head=traverse_input()
print_ll(head)
rev_ll=reverse(head)
print_ll(rev_ll)
```

```python
class Stack:
    def __init__(self,n):
        self.__data=[]
        self.__count=0
        self.limit=n
    def push(self,data):
        if self.__count<self.limit:
            self.__data.append(data)
            self.__count+=1
        else:
            print("Stack is full.")
            
    def pop(self):
        if self.__count<=0:
            return "Stack is empty can't operate the pop function."
        ele=self.__data.pop()
        self.__count-=1
        return ele
    def top(self):
        if self.__count<=0:
            return "Stack is empty."
        return self.__data[-1]
    def size(self):
        return self.__count
    def isEmpty(self):
        return self.__count==0

s=Stack(6)
s.push(1)
s.push(2)
s.push(3)
s.push(4)
s.push(5)
s.push(6)
s.push(7)
print("Top",s.top())
while s.isEmpty()!=True:
    print(s.pop())
print(s.pop())
```

```python
def check_paranthesis(string):
    if len(string)==0:
        return True
    s=[]
    for i in string:
        if i=='[' or i=='(' or i=='{':
            s.append(i)
        elif len(s)!=0 and i=='}' and s[-1]=='{':
            s.pop()
        elif len(s)!=0 and i==')' and s[-1]=='(':
            s.pop()
        elif len(s)!=0 and i==']' and s[-1]=='[':
            s.pop()
        else:
            s.append(i)
    if len(s)==0:
        return True
    return False

    
string=input()
check_paranthesis(string)
```

```python
class stack:
    def __init__(self,s):
        self.data=[]
        self.length=s
        self.count=0
    def push(self,ele):
        if self.count<self.length:
            self.data.append(ele)
            self.count+=1
        else:
            print("The stack is full cant operate the push operation.")
    def pop(self):
        if self.count>0:
            ele=self.data.pop()
            self.count-=1
            return ele

        else:
            print("Stack is empty cant pop")

    def top(self):
        if self.count>0:
            return self.data[-1]
        else:
            print("Stack is empty")
        
    def size(self):
        return self.count
    def isEmpty(self):
        return self.size()==0

s=stack(6)
s.push(1)
s.push(2)
s.push(3)
s.push(4)
s.push(5)
s.push(6)
print("Top",s.top())
while s.isEmpty()!=True:
    print("Pop",s.pop())
print("Top",s.top())
print("pop",s.pop())
```

```python
class stack:
    def __init__(self,limit):
        self.count=0
        self.s=[]
        self.limit=limit
    
    def push(self,data):
        if self.count<self.limit:
            self.s.append(data)
            self.count+=1
        else:
            print("stack overflow can't append")
            return 
    def pop(self):
        if self.count>0:
            ele=self.s[-1]
            self.s.pop()
            self.count-=1
            return ele
        else:
            return ("Stack is empty so cant perform pop functionality.")
            
    def top(self):
        if self.count>0:
            return self.s[-1]
        else:
            return "Stack is empty so not any top element."
        
    def isempty(self):
        return self.count==0
    def length(self):
        return self.count
s=stack(9)
s.push(1)
s.push(2)
s.push(3)
s.push(4)
s.push(5)
print("Length",s.length())
print("Top",s.top())
while s.isempty()!=True:
    print(s.pop())
print(s.pop())
print(s.top())
    

        
```

```python
class queue:
    def __init__(self,limit):
        self.count=0
        self.l=[]
        self.limit=limit
    
    def enqueue(self,data):
        if self.count<self.limit:
            self.l.append(data)
            self.count+=1
        else:
            return "Queue is full cant enqueue"
        
    def dequeue(self):
        if self.count>0:
            ele=self.l[0]
            self.l.pop(0)
            self.count-=1
            return ele
        else:
            return "Queue is empty cany dequeue."
        
    def front(self):
        if self.count>0:
            return self.l[0]
        else:
            return "Queue is empty no front"
    
    def size(self):
        return self.count
    
    def isempty(self):
        return self.count==0
    
q=queue(9)
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
q.enqueue(5)
print("size",q.size())
print("front",q.front())
while q.isempty()!=True:
    print(q.dequeue())
print(q.dequeue())
print(q.front())
```
