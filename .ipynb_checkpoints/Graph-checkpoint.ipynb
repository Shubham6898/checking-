```python
# BFS(Breadth First Search): --> just like Level wise traverse
import queue
class Graph:
    def __init__(self,novertices):
        self.no_of_vertices=novertices
        self.adj_matrix=[[0 for j in range(novertices)] for i in range(novertices)]

    def addedge(self,v1,v2):
        self.adj_matrix[v1][v2]=1
        self.adj_matrix[v2][v1]=1
    
    def remove_edge(self,v1,v2):
        if self.contain_edge[v1][v2]==False:
            return
        self.adj_matrix[v1][v2]=0
        self.adj_matrix[v2][v1]=0

    def contain_edge(self,v1,v2):
        if self.adj_matrix[v1][v2]==0:
            return False
        return True
    def BFS(self):
        q=queue.Queue()
        q.put(0)
        visited=[False for i in range(self.no_of_vertices)]
        visited[0]=True
        while q.empty()!=True:
            current=q.get()
            print(current)
            for i in range(self.no_of_vertices):
                if self.adj_matrix[current][i]>0 and visited[i]==False:
                    q.put(i)
                    visited[i]=True
    def __str__(self):
        return str(self.adj_matrix)

g=Graph(6)
g.addedge(0,3)
g.addedge(0,1)

g.addedge(1,2)
g.addedge(2,3)
g.addedge(2,5)
g.addedge(3,4)
g.addedge(4,5)
g.BFS()
print('-------2ND--------')
g2=Graph(5)
g2.addedge(0,1)
g2.addedge(1,3)
g2.addedge(2,4)
g2.addedge(0,2)
g2.BFS()
```

# Implementation of Graph using Adjacency List : 
An array of lists is used. The size of the array is equal to the number of vertices. Let the array be an array[]. An entry array[i] represents the list of vertices adjacent to the ith vertex. This representation can also be used to represent a weighted graph. The weights of edges can be represented as lists of pairs. Following is the adjacency list representation of the above graph. 


```python
class AdjNode:
    def __init__(self,data): # yeh every vertices ke liye class hai
        self.vertex=data
        self.next=None

class Graph:
    def __init__(self,novertices):
        self.no_of_vertices=novertices
        self.adj_list=[None]*self.no_of_vertices # yeh humari adjacent list hogyi

    def add_edge(self,v1,v2): # v1 is the starting vertex and v2 is the ending vertex   
        node=AdjNode(v1)
        node.next=self.adj_list[v2]
        self.adj_list[v2]=node

        node=AdjNode(v2)
        node.next=self.adj_list[v1]
        self.adj_list[v1]=node

    def print_graph(self):
        for  i in range(self.no_of_vertices):
            print(i,end='-->')
            temp=self.adj_list[i]
            while temp:
                print(temp.vertex,end=' ')
                temp=temp.next
            print()
graph=Graph(5)

graph.add_edge(0, 1)
graph.add_edge(0, 4)
graph.add_edge(1, 2)
graph.add_edge(1, 3)
graph.add_edge(1, 4)
graph.add_edge(2, 3)
graph.add_edge(3, 4)

graph.print_graph()
a=(graph.adj_list)
for i in a:
    print(i.vertex)
```

```python
"""
A Python program to demonstrate the adjacency
list representation of the graph
"""

# A class to represent the adjacency list of the node


class AdjNode:
	def __init__(self, data):
		self.vertex = data
		self.next = None


# A class to represent a graph. A graph
# is the list of the adjacency lists.
# Size of the array will be the no. of the
# vertices "V"
class Graph:
	def __init__(self, vertices):
		self.V = vertices
		self.graph = [None] * self.V

	# Function to add an edge in an undirected graph
	def add_edge(self, src, dest):
		# Adding the node to the source node
		node = AdjNode(dest)
		node.next = self.graph[src]
		self.graph[src] = node

		# Adding the source node to the destination as
		# it is the undirected graph
		node = AdjNode(src)
		node.next = self.graph[dest]
		self.graph[dest] = node

	# Function to print the graph
	def print_graph(self):
		for i in range(self.V):
			print("Adjacency list of vertex {}\n head".format(i), end="")
			temp = self.graph[i]
			while temp:
				print(" -> {}".format(temp.vertex), end="")
				temp = temp.next
			print(" \n")


# Driver program to the above graph class
if __name__ == "__main__":
	V = 5
	graph = Graph(V)
	graph.add_edge(0, 1)
	graph.add_edge(0, 4)
	graph.add_edge(1, 2)
	graph.add_edge(1, 3)
	graph.add_edge(1, 4)
	graph.add_edge(2, 3)
	graph.add_edge(3, 4)

	graph.print_graph()

# This code is contributed by Kanav Malhotra

a=(graph.graph)
for i in a:
    print(i.vertex)
# This code is contributed by Kanav Malhotra
```

```python
#BFS:
import queue
class Graph:
    def __init__(self,no_vertices):
        self.no_of_vertices=no_vertices
        self.adj_matrix=[[0]*no_vertices for i in range(self.no_of_vertices)]
    
    def add_edge(self,v1,v2):
        self.adj_matrix[v1][v2]=1
        self.adj_matrix[v2][v1]=1
    def remove_edge(self,v1,v2):
        if self.contain_edge(v1,v2)==True:
            self.adj_matrix[v1][v2]=0
            self.adj_matrix[v2][v1]=0
        else:
            return
    
    def contain_edge(self,v1,v2):
        return  self.adj_matrix[v1][v2]!=0
    
    def BFS(self,sv):
        q=queue.Queue()
        visited=[False for i in range(self.no_of_vertices)]
        q.put(sv)
        visited[sv]=True
        while q.empty()!=True:
            current_node=q.get()
            print(current_node)
            for i in range(self.no_of_vertices):
                if self.adj_matrix[current_node][i]!=0 and visited[i]!=True:
                    q.put(i)
                    visited[i]=True
    

    def __str__(self):
        return str(self.adj_matrix)


g=Graph(6)
g.add_edge(0,1)
g.add_edge(1,2)
g.add_edge(2,3)
g.add_edge(0,3)
g.add_edge(3,4)
g.add_edge(2,5)
g.add_edge(4,5)
g.BFS(0)     
```

```python
# adjacent_list representation:

class adjacent_node:
    def __init__(self,data):
        self.data=data
        self.next=None

class Graph:
    def __init__(self,no_vertices):
        self.no_of_vertices=no_vertices
        self.adj_list=[None]*no_vertices
    
    def add_edge(self,v1,v2):
        # dest to source
        node=adjacent_node(v2)
        node.next=self.adj_list[v1]
        self.adj_list[v1]=node

        # source to dest
        node=adjacent_node(v1)
        node.next=self.adj_list[v2]
        self.adj_list[v2]=node

    def print_graph(self):
        for i in range(self.no_of_vertices):
            print("Adjacency list of vertex",i, end=" ")
            temp = self.adj_list[i]
            while temp:
                print(" -> ",temp.data, end="")
                temp = temp.next
            print(" \n")

graph = Graph(5)
graph.add_edge(0, 1)
graph.add_edge(0, 4)
graph.add_edge(1, 2)
graph.add_edge(1, 3)
graph.add_edge(1, 4)
graph.add_edge(2, 3)
graph.add_edge(3, 4)
graph.print_graph()

'''
Dekh idhar jo tune adjacent list bnai hai usme kya ho rha hai ki uss list mei har ek node ke liye ek linked lised saved hai as you
can see in the output down toh chakkr yeh hai ki jab tu jaye ga ek edge ko delete karne ke liye tab terko yeh dono vertices jinke bich mei
edge hai unki respective linked list ko alg alg traverse karna pdega aur phir deletion ka code on linked list use karna pdega
'''
```

```python
# has Path:

class Graph:
    def __init__(self,no_vertices):
        self.no_of_vertices=no_vertices
        self.adj_matrix=[[0]*self.no_of_vertices for i in range(no_vertices)]
    
    def add_edge(self,v1,v2):
        self.adj_matrix[v1][v2]=1
        self.adj_matrix[v2][v1]=1

    def remove_edge(self,v1,v2):
        if self.contain_edge(v1,v2)==True:
            self.adj_matrix[v1][v2]=0
            self.adj_matrix[v2][v1]=0

        else:
            return 
    def conatin_edge(self,v1,v2):
        return self.adj_matrix[v1][v2]!=0
    
    def __str__(self):
        return str(self.adj_matrix)
    
    def has_path_helper(self,v1,v2,visited):
        if self.adj_matrix[v1][v2]==1:
            return True
        visited[v1]=True
        for i in range(self.no_of_vertices):
            if self.adj_matrix[v1][i]!=0 and visited[i]!=True:
                if self.has_path_helper(i,v2,visited):
                    return True
        return False

    def has_path(self,v1,v2):
        visited=[False for i in range(self.no_of_vertices)]
        if self.has_path_helper(v1,v2,visited):
            return True
        return False

g=Graph(9)
g.add_edge(0, 1)
g.add_edge(0, 3)
g.add_edge(1, 2)
g.add_edge(2, 3)
g.add_edge(2, 5)
g.add_edge(3, 4)
g.add_edge(4, 5)
g.add_edge(6,7)
g.has_path(8,7)    
```

```python
class graph:
    def __init__(self,nodes):
        self.node_list=nodes
        self.adj_dict={}
        for i in nodes:
            self.adj_dict[i]=[]

    def add_edge(self,v1,v2):
        self.adj_dict[v1].append(v2)
        self.adj_dict[v2].append(v1)
    def remove_edge(self,v1,v2):
        self.adj_dict[v1].remove(v2)
        self.adj_dict[v2].rempve(v1)
    def print_graph(self):
        for i in self.node_list:
            print(i,"-->",self.adj_dict[i])
    
    def has_path_helper(self,v1,v2,visited):
        if v2 in self.adj_dict[v1]:
            return True
        visited[v1]=True
        for i in self.adj_dict[v1]:
            if visited[i]==False:
                if self.has_path_helper(i,v2,visited):
                    return True
        return False

    def has_path(self,v1,v2):
        visited={}
        for i in self.node_list:
            visited[i]=False
        return self.has_path_helper(v1,v2,visited)
nodes=[0,1,2,3,4,5,6,7,8]
g=graph(nodes)
g.add_edge(0, 1)
g.add_edge(0, 3)
g.add_edge(1, 2)
g.add_edge(2, 3)
g.add_edge(2, 5)
g.add_edge(3, 4)
g.add_edge(4, 5)
g.add_edge(6,7)
g.has_path(1,4) 
```
