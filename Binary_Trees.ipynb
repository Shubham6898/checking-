```python
import queue
class Binary_tree_node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None


def print_tree_detailed(root):
    if root is None:
        return None
    print(root.data,end=': ')
    if root.left!=None:
        print('L: ',root.left.data,end=', ')
    if root.right!=None:
        print('R: ',root.right.data)
    print()
    print_tree_detailed(root.left)
    print()
    print_tree_detailed(root.right)

def build_tree_pre_ino(preo,ino):
    if len(ino)==0:
        return None
    root_data=preo[0]
    root=Binary_tree_node(root_data)
    root_index_in_ino=-1
    for i in range(len(ino)):
        if ino[i]==root_data:
            root_index_in_ino=i
            break
        
    if root_index_in_ino==-1:
        return None
    left_ino=ino[:root_index_in_ino]
    right_ino=ino[root_index_in_ino+1:]
    
    left_preo=preo[1:len(left_ino)+1]
    right_preo=preo[len(left_ino)+1:]
    
    left_child=build_tree_pre_ino(left_preo,left_ino)
    right_child=build_tree_pre_ino(right_preo,right_ino)
    root.left=left_child
    root.right=right_child
    return root

preo=[1,2,4,5,3,6,7]
ino=[4,2,5,1,6,3,7]
print_tree_detailed(build_tree_pre_ino(preo,ino))
    
```
