```python
class Stack:
    def __init__(self,limit):
        self.arr=[]
        self.limit=limit
        self.count=0
        
    def push(self,data):
        if self.count<self.limit:
            self.arr.append(data)
            self.count+=1
        else:
            print("Stack is full for pushing.")
    
    def pop(self):
        if self.count>0:
            ele=self.arr.pop()
            self.count-=1
            return ele
        else:
            return 'Stack is empty.'
    def top(self):
        if self.count>0:
            return self.arr[-1]
        return 'Stack is empty no top.'
        
    def size(self):
        return self.count
    def isEmpty(self):
        return self.count==0
    
s=Stack(9)
s.push(1)
s.push(2)
s.push(3)
s.push(4)
s.push(5)
print("Length",s.size())
print("Top",s.top())
while s.isEmpty()!=True:
    print(s.pop())
print(s.pop())
print(s.top())
```

```python
# bubble sort:
def bubble_sort(arr,n):
    for i in range(n-1):
        for j in range(n-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr

n=int(input())
arr = [64, 34, 25, 12, 22, 11, 90]
print(bubble_sort(arr,n))
            
    
```

```python
def bubble_sort_imp(arr,n):
    count=0
    for i in range(n-1):
        swap=False
        count+=1
        for j in range(n-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                swap=True
        if swap==False:
            break
    return [arr,count]

n=int(input())
arr = [11, 12, 22, 25, 34, 64, 90]
print(bubble_sort_imp(arr,n))
```

```python
def selection_sort(arr,n):
    for i in range(n):
        min_index=i
        for j in range(i+1,n):
            if arr[min_index]>arr[j]:
                min_index=j
        arr[min_index],arr[i]=arr[i],arr[min_index]
    return arr

n=int(input())
arr = [64, 34, 25, 12, 22, 11, 90]
print(selection_sort(arr,n))
        
```

```python
def insertion_sort(arr,n):
    for i in range(1,n):
        temp=arr[i]
        j=i-1
        while j>=0 and temp<arr[j]:
            arr[j+1]=arr[j]
            j-=1
        arr[j+1]=temp
    return arr

n=int(input())
arr = [64, 34, 25, 12, 22, 11, 90]
print(insertion_sort(arr,n))
```
